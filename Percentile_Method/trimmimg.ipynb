```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('weight-height.csv')
df.sample(5)
```

```python
df['Height'].describe()
```

```python
sns.distplot(df['Height'])
```

```python
sns.boxplot(df['Height'])
```

```python
pr_max = df['Height'].quantile(0.99)
pr_min = df['Height'].quantile(0.01)
print("Pr_max =",pr_max)
print("Pr_min =",pr_min)
```

```python
new_df = df[(df['Height']>pr_min) & (df['Height']<pr_max)]
new_df.sample(5)
```

```python
new_df['Height'].describe()
```

```python
sns.distplot(new_df['Height'])
```

```python
sns.boxplot(new_df['Height'])
```
