```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('weight-height.csv')
df.sample(5)
```

```python
df['Height'].describe()
```

```python
sns.distplot(df['Height'])
```

```python
sns.boxplot(df['Height'])
```

```python
pr_max = df['Height'].quantile(0.99)
pr_min = df['Height'].quantile(0.01)
print("Pr_max =",pr_max)
print("Pr_min =",pr_min)
```

```python
df['Height'] = np.where(df['Height']>pr_max,
                  pr_max,
                  np.where(df['Height']<pr_min,
                           pr_min,
                           df['Height']))
```
